# Procesamiento de imágenes en python

```python
import matplotlib.pyplot as plt
import numpy as np
from keras.preprocessing import image as image_utils
import cv2
import imutils
```

# Lectura de las imágenes

```python
imagen = cv2.imread('cat.jpg')
imagen = cv2.cvtColor(imagen, cv2.COLOR_RGB2BGR)
plt.imshow(imagen)
plt.axis('off')
plt.show()
```

```python
imagen.shape
```

```python
imagen2 = image_utils.load_img('cat.jpg')
plt.imshow(imagen2, cmap='Accent')
plt.show()
imagen2 = image_utils.img_to_array(imagen2)
```

```python
img_gray = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap='gray')
plt.show()
```

```python
img_gray.shape
```

```python
img_hsv = cv2.cvtColor(imagen2, cv2.COLOR_BGR2HSV)
plt.imshow(img_hsv, cmap='hsv')
plt.show()
```

```python
imagesize = cv2.resize(imagen2, (1200,800), interpolation = cv2.INTER_CUBIC)
imagesize = image_utils.array_to_img(imagesize)
plt.imshow(imagesize, cmap='Accent')
plt.show()
```

```python
alto, ancho, canales = imagen2.shape
M = cv2.getRotationMatrix2D((ancho/2, alto/2), 45, 1)
imgrotate = cv2.warpAffine(imagen2, M, (ancho, alto))
plt.imshow(image_utils.array_to_img(imgrotate), cmap='Accent')
plt.show()
```

```python
imgrotate2 = imutils.rotate_bound(imagen2, 45)
plt.imshow(image_utils.array_to_img(imgrotate2), cmap='Accent')
plt.show()
```

```python
gaussian = np.zeros(imagen2.shape)
cv2.randn(gaussian, 0.0, 40)
imgruido = imagen2 + gaussian
plt.imshow(image_utils.array_to_img(imgruido), cmap='Accent')
plt.show()
```

```python
imglimpia = cv2.GaussianBlur(imgruido, (5,7), 0)
plt.imshow(image_utils.array_to_img(imglimpia), cmap='Accent')
plt.show()
```

```python
img_canny = cv2.Canny(img_gray.astype('uint8'), 150,200)
plt.imshow(img_canny, cmap='gray')
plt.show()
```
